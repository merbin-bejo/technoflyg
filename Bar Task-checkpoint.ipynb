```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
tec=np.array(["SIA[28]","TSbMH[30]","Proposed HV-AL"])
et=np.array([20,100,15])

plt.xlabel("Techniques")
plt.ylabel("Execution time")
plt.title("Comparison performance of execution time")
plt.rcParams['figure.figsize'] = [5, 6]

plt.bar(tec,et,height=0.3)
```

```python
c=['Memory usage (%)']
a=np.array(['TSbMH[30]', 'MMPCM[34]', 'PROPOSED HV-ALM'])
b =np.array([51,50,30])
plt.rcParams['figure.figsize'] = [6, 4]
plt.barh(a, b, color ='#301F21',
        height = 0.4) 
plt.xlabel("Memory usage(%)",color="#4D6A06",fontsize=18)
plt.title("Comparison performance of memory usage ",color="#79923C")
plt.legend(c)
```
